```python
import json
from pathlib import Path
import re

# Enabling tracing to LangSmith
from langsmith import traceable
#from langsmith.tracing import tracing_context

#with tracing_context(enabled=True, project_name="DEIN_PROJEKTNAME"):answer = ask_with_articles(...)



history_path = Path("kontextverlauf.json")
if not history_path.exists():
    history_path.write_text("[]")


def save_to_history(question, answer):
    with open(history_path, "r+", encoding="utf-8") as file:
        history = json.load(file)
        history.append({"frage": question, "antwort": answer})
        file.seek(0)
        json.dump(history, file, ensure_ascii=False, indent=2)
@traceable(name="ask_with_articles")
def ask_with_articles(qa, retriever, question, history=[], num_history=3, num_docs=2):
    # Step 1: Load latest Q&A from history
    context_history = "\n".join([
        f"Frage: {item['frage']}\nAntwort: {item['antwort']}"
        for item in history[-num_history:]
    ])

    # Step 2: Retrieve relevant document chunks
    retrieved_docs = retriever.get_relevant_documents(question)


    doc_context = "\n".join([doc.page_content for doc in retrieved_docs[:num_docs]])

    #print(f"Frage test:{question} Context:{doc_context}")



    prompt = (
    f"Vorherige Fragen und Antworten (falls relevant):\n{context_history}\n\n"
    f"Kontext aus den Dokumenten:\n{doc_context}\n\n"
    f"Neue Frage: {question}"
)


    # Step 4: Send to LLM
    response = qa.invoke({"query": prompt})


    return response
def load_history():
    if os.path.exists(history_path):
        with open(history_path, "r", encoding="utf-8") as f:
            return json.load(f)
    return []
```

```python
import requests
from bs4 import BeautifulSoup
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI
from langchain.chains import RetrievalQA

# 🔗 List of URLs (5 articles)
urls = [
    "https://www.dw.com/de/nach-debakel-im-ersten-wahlgang-merz-neuer-bundeskanzler-deutschland-kanzler-regierung-cdu-v2/a-72449047",
    "https://www.deutschlandfunk.de/regierung-von-bundeskanzler-merz-uebernimmt-amtsgeschaefte-100.html",
    "https://www.bundestag.de/dokumente/textarchiv/2025/kw19-de-kanzlerwahl-1062470",
    "https://www.zdf.de/nachrichten/politik/deutschland/merz-kanzler-bundeskanzler-wahl-bundestag-cdu-100.html",
    "https://www.zdf.de/nachrichten/thema/papst-franziskus-214.html",
    "https://www.tagesschau.de/ausland/europa/papst-leo-100.html",
    "https://www.tagesschau.de/ausland/asien/syrien-kaempfe-alawiten-100.html",
    "https://uswahl.lpb-bw.de/usa-wahlergebnisse-2024",
    "https://www.mdr.de/nachrichten/sachsen-anhalt/magdeburg/magdeburg/chronik-anschlag-weihnachtsmarkt-tote-opfer-taeter-motiv-104.html",
    "https://www.diplomatie.gouv.fr/de/landerinformationen/nordafrika-und-mittlerer-osten/syrien/neuigkeiten/article/syrien-sturz-des-regimes-von-baschar-al-assad-8-dezember-2024",
    "https://de.wikipedia.org/wiki/Anschlag_auf_den_Magdeburger_Weihnachtsmarkt",
    "https://de.wikipedia.org/wiki/Pr%C3%A4sident_der_Vereinigten_Staaten"
]

# 📄 Document preparation
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
#print("Chunks :",len(splitter))
all_docs = []

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    text = soup.get_text()
    paragraphs = soup.find_all("p")
    text = "\n".join([para.get_text() for para in paragraphs if para.get_text()])

    docs = splitter.create_documents([text])
    docs = splitter.create_documents([text])

    #print(docs)
    all_docs.extend(docs)
print(f"Total number of chunks: {len(all_docs)}")

# 🔎 Embedding + Vector Store
embedding = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
vectordb = Chroma.from_documents(documents=all_docs, embedding=embedding, persist_directory="db")
vectordb.persist()
print("Alle Artikel erfolgreich verarbeitet.")

# 🤖 QA Chain Setup
retriever = vectordb.as_retriever()
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.3,convert_system_message_to_human=True)
qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=retriever)




for q in test_questions:

    #antwort = qa_chain.invoke({"query": q})
    history = load_history()
    answer = ask_with_articles(qa_chain, retriever, q, history)
    print("\nFrage:", q)
    print("Antwort:", answer['result'])
    print("-" * 60)


# ❓ Ask 5 Questions
for _ in range(5):
    history = load_history()
    question = input("\nStelle eine Frage ('exit' zum Beenden): ")
    if question.lower() == "exit":
        break
    #history = load_history()  # list of {"frage": "...", "antwort": "..."}
    answer = ask_with_articles(qa_chain, retriever, question, history)
    #answer = qa_chain.run(question)
    print(answer['result'])
    save_to_history(question, answer['result'])

```
